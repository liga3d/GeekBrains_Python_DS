# Задание 2

```python
t = np.linspace(0, 10, 51)
t
```

```python
f = np.cos(t)
f
```

```python
text_dict = {'fontsize': '12', 'color': '#404040', 'family': 'Calibri'}

plt.axis([0.5, 9.5, -2.5, 2.5])

plt.plot(t, f, label='cos(t)', color='darkgreen')

plt.title('Графики f(t)', fontdict=text_dict)
plt.xlabel('Значения t', fontdict=text_dict)
plt.ylabel('Значения f', fontdict=text_dict)

plt.grid(color='#E0E0E0')
plt.legend()
plt.show()
```

# Задание 3

```python
x = np.linspace(-3, 3, 51)
x
```

```python
y1 = x ** 2
y1
```

```python
y2 = 2 * x + 0.5
y2
```

```python
y3 = -3 * x - 1.5
y3
```

```python
y4 = np.sin(x)
y4
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1)
ax1.set_title('График y1')
ax1.set_xlim([-5, 5])

ax2.plot(x, y2)
ax2.set_title('График y2')

ax3.plot(x, y3)
ax3.set_title('График y3')

ax4.plot(x, y4)
ax4.set_title('График y4')

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```
